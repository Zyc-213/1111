```python
import os
from shapely.wkt import loads
from shapely.geometry import Polygon

# Define file path to precomputed FCI data chunk map in WKT format
wkt_file_path = "FCI_chunks.wkt"

if not os.path.exists(wkt_file_path):
    raise FileNotFoundError(f"File {wkt_file_path} not found. Make sure it is in the repository.")

# Load WKT chunk footprints
with open(wkt_file_path, "r") as file:
    wkt_data = file.readlines()

# Parse chunk polygons from WKT
chunk_polygons = {}
for line in wkt_data:
    chunk_id, wkt_poly = line.strip().split(',', 1)  # Extract chunk ID
    chunk_polygons[chunk_id] = loads(wkt_poly)

print(f"Loaded {len(chunk_polygons)} chunk footprints from WKT file.")
```

```python
#Convert user ROI to a Shapely Polygon
roi_polygon = Polygon([
    (user_roi["lon_min"], user_roi["lat_min"]),
    (user_roi["lon_min"], user_roi["lat_max"]),
    (user_roi["lon_max"], user_roi["lat_max"]),
    (user_roi["lon_max"], user_roi["lat_min"])
])

# Find chunks that intersect with ROI
relevant_chunks = []
for chunk_id, chunk_poly in chunk_polygons.items():
    if roi_polygon.intersects(chunk_poly):
        relevant_chunks.append(chunk_id)

print(f"Found {len(relevant_chunks)} chunks intersecting the ROI: {relevant_chunks}")
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from shapely.geometry import LineString
from pyproj import CRS

wgs84_crs = CRS.from_proj4("+proj=longlat +ellps=WGS84 +datum=WGS84 +no_defs")

# Convert chunk polygons to a GeoDataFrame
gdf_chunks = gpd.GeoDataFrame({"chunk_id": list(chunk_polygons.keys()), "geometry": list(chunk_polygons.values())}, crs=wgs84_crs)
# Plot setup
fig, ax = plt.subplots(figsize=(12, 12), subplot_kw={"projection": ccrs.PlateCarree()})
ax.set_extent([-90, 90, -90, 90])
ax.coastlines("50m", linewidth=0.25)
ax.add_feature(cfeature.LAND, facecolor="lightgray", edgecolor="black", linewidth=0.25)

# Plot chunks with labels
for i, row in gdf_chunks.iterrows():
    chunk_id, chunk_poly = row["chunk_id"], row["geometry"]
    if not chunk_poly.is_valid: continue
    ax.fill(*chunk_poly.exterior.xy, color=plt.cm.tab20.colors[i % 20], alpha=0.25, transform=ccrs.PlateCarree())

    # Label position inside polygon
    center_x = (chunk_poly.bounds[0] + chunk_poly.bounds[2]) / 2
    vertical_line = LineString([(center_x, chunk_poly.bounds[1]), (center_x, chunk_poly.bounds[3])])
    label_y = vertical_line.intersection(chunk_poly).centroid.y
    ax.text(center_x, label_y, chunk_id, fontsize=6, transform=ccrs.PlateCarree(), ha="center", va="center")

# Highlight ROI
ax.plot(*roi_polygon.exterior.xy, color="red", linewidth=1, linestyle="dashed", transform=ccrs.PlateCarree())
plt.title("MTG Chunk coverage extent and user ROI")
plt.show()
```

```python
import os
import shutil

def download_chunks_in_time_window(selected_collection, dtstart, dtend, chunk_ids, output_dir):
    """
    Search for products in the given time window, download relevant .nc entries and trailer chunk (0041).
    All files will be saved to the specified output folder.
    """
    # 创建目标文件夹，不存在则新建
    os.makedirs(output_dir, exist_ok=True)

    # Always ensure trailer chunk "0041" is included
    chunk_ids.append("0041")
    # 去重，防止重复下载0041
    chunk_ids = list(set(chunk_ids))

    chunk_patterns = [f"_{cid}.nc" for cid in chunk_ids]

    # Products in time window
    products = selected_collection.search(dtstart=dtstart, dtend=dtend)
    products_list = list(products)
    print(f"Found {len(products_list)} matching timestep(s).")

    # Filter relevant entries
    for product in products_list:
        for entry in product.entries:
            if any(pattern in entry for pattern in chunk_patterns):
                try:
                    with product.open(entry=entry) as fsrc:
                        local_filename = os.path.basename(fsrc.name)
                        # 拼接完整保存路径：文件夹 + 文件名
                        full_save_path = os.path.join(output_dir, local_filename)
                        print(f"Downloading file {local_filename}...")
                        with open(full_save_path, 'wb') as fdst:
                            shutil.copyfileobj(fsrc, fdst)
                        print(f"Saved file -> {full_save_path}")
                except Exception as e:
                    print(f"Error downloading {entry}: {e}")
```

```python
import numpy as np
import pandas as pd
from scipy.interpolate import griddata
import mlcast_datasets
import os

# 1. 加载卫星固定网格
sat_dir = r"fci_data_train/fci_npy_output"
lon2d = np.load(f"{sat_dir}/lon_grid.npy")
lat2d = np.load(f"{sat_dir}/lat_grid.npy")
y_size, x_size = lon2d.shape
sat_times = np.load(f"{sat_dir}/times.npy", allow_pickle=True)

# 2. 读取DMI雷达在线数据集（Colab专用，本地不能跑）
cat = mlcast_datasets.open_catalog()
radar_all = cat.precipitation.dmi_10_minutes.to_dask()

# 截取卫星前后时间段
t_start = pd.to_datetime(sat_times[0]) - pd.Timedelta("10min")
t_end = pd.to_datetime(sat_times[-1]) + pd.Timedelta("10min")
radar_window = radar_all.sel(time=slice(t_start, t_end)).compute()

# 雷达坐标点
radar_lon = radar_window.lon.values
radar_lat = radar_window.lat.values
radar_points = np.column_stack([radar_lon.ravel(), radar_lat.ravel()])
target_points = np.column_stack([lon2d.ravel(), lat2d.ravel()])

# 输出雷达标签文件夹
radar_out = "/content/radar_label_npy"
os.makedirs(radar_out, exist_ok=True)
match_idx_list = []

# 遍历每一张卫星时间帧
for idx, sat_t in enumerate(sat_times):
    print(f"匹配卫星时间：{sat_t}")
    # 计算雷达时间差
    dt_diff = np.abs(radar_window.time.values - np.datetime64(sat_t))
    best_r_idx = dt_diff.argmin()

    # 差值大于5分钟直接跳过
    if dt_diff[best_r_idx] > np.timedelta64(5, "m"):
        print(f"{sat_t} 距离最近雷达超过5分钟，跳过")
        continue
    match_idx_list.append(idx)

    # 取出对应雷达反射率
    radar_dbz_raw = radar_window.dbz.isel(time=best_r_idx).values
    radar_flat = radar_dbz_raw.ravel()

    # 插值到卫星网格
    radar_interp = griddata(radar_points, radar_flat, target_points, method="linear")
    radar_2d = radar_interp.reshape(y_size, x_size).astype(np.float32)
    radar_2d = np.nan_to_num(radar_2d, nan=-30)

    # 保存单帧雷达 (1, y, x)
    np.save(f"{radar_out}/radar_{idx:03d}.npy", radar_2d[None, :, :])

# 保存有效配对序号
np.save(f"{sat_dir}/valid_match_idx.npy", np.array(match_idx_list))
print(f"\n✅ 时空配对完成，有效样本总数：{len(match_idx_list)}")
```

```python
# ========== 第一步：检查文件路径 ==========
import os
print("当前目录文件：")
print(os.listdir("/content"))

SAT_DIR = "/content/fci_data_train/fci_npy_output"
RADAR_DIR = "/content/radar_label_npy"

print(f"\n卫星文件数：{len(os.listdir(SAT_DIR))}")
print(f"雷达文件数：{len(os.listdir(RADAR_DIR))}")
```

```python
# ========== 第二步：导入库 + 定义数据集 ==========
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import matplotlib.pyplot as plt

class SatRadarDataset(Dataset):
    def __init__(self, sat_dir, radar_dir):
        self.sat_dir = sat_dir
        self.radar_dir = radar_dir
        self.valid_ids = np.load(f"{sat_dir}/valid_match_idx.npy")

    def __len__(self):
        return len(self.valid_ids)

    def __getitem__(self, i):
        raw_idx = self.valid_ids[i]  # 这里是0/1/2...
        print(raw_idx)
        real_file_num = raw_idx + 1  # 0→1，匹配 frame_0001

        sat = np.load(f"{self.sat_dir}/frame_{real_file_num:04d}.npy").astype(np.float32)
        # 雷达同步+1，统一4位命名
        radar = np.load(f"{self.radar_dir}/radar_{raw_idx:03d}.npy").astype(np.float32)

        sat = np.clip((sat - 200) / 200, 0, 1)
        radar = np.clip((radar + 30) / 90, 0, 1)
        return torch.from_numpy(sat), torch.from_numpy(radar)
```

```python
class DoubleConv(nn.Module):
    def __init__(self, in_c, out_c):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv2d(in_c, out_c, 3, padding=1),
            nn.BatchNorm2d(out_c), nn.ReLU(inplace=True),
            nn.Conv2d(out_c, out_c, 3, padding=1),
            nn.BatchNorm2d(out_c), nn.ReLU(inplace=True)
        )
    def forward(self, x):
        return self.net(x)

class Sat2RadarUNet(nn.Module):
    def __init__(self, in_channels=5):
        super().__init__()
        # 编码器
        self.in_conv = DoubleConv(in_channels, 32)   # 5→32
        self.down1 = DoubleConv(32, 64)              # 32→64
        self.down2 = DoubleConv(64, 128)             # 64→128

        # 解码器（拼接后通道 = 上采样通道 + skip通道）
        self.up1 = DoubleConv(128 + 64, 64)          # 128+64=192 → 64
        self.up2 = DoubleConv(64 + 32, 32)           # 64+32=96 → 32

        self.out = nn.Conv2d(32, 1, 1)
        self.pool = nn.MaxPool2d(2)
        self.up = nn.Upsample(scale_factor=2, mode="bilinear")

    def forward(self, x):
        # 下采样
        x1 = self.in_conv(x)          # 32通道
        x2 = self.down1(self.pool(x1)) # 64通道
        x3 = self.down2(self.pool(x2)) # 128通道

        # 上采样 + 跳跃连接
        x = self.up1(torch.cat([self.up(x3), x2], dim=1))  # 128+64=192 → 64
        x = self.up2(torch.cat([self.up(x), x1], dim=1))   # 64+32=96 → 32

        return self.out(x)
```

```python
# ========== 第四步：训练 ==========
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"\n训练设备：{device}")

dataset = SatRadarDataset(SAT_DIR, RADAR_DIR)
train_loader = DataLoader(dataset, batch_size=2, shuffle=True)
print(f"训练样本数：{len(dataset)}")

model = Sat2RadarUNet().to(device)
loss_fn = nn.MSELoss()
opt = torch.optim.Adam(model.parameters(), lr=1e-4)

print("\n开始训练...")
for epoch in range(100):
    total_loss = 0
    for x, y in train_loader:
        x, y = x.to(device), y.to(device)
        pred = model(x)
        loss = loss_fn(pred, y)
        opt.zero_grad()
        loss.backward()
        opt.step()
        total_loss += loss.item()
    if (epoch+1) % 5 == 0:
        print(f"Epoch {epoch+1:2d} | MSE Loss: {total_loss/len(train_loader):.5f}")
```

```python
model.eval()
x_sample, y_sample = next(iter(train_loader))
with torch.no_grad():
    pred = model(x_sample.to(device)).cpu()

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
axes[0].imshow(x_sample[0, 3], cmap="gray")
axes[0].set_title("FCI Water Vapor (Input)")
axes[1].imshow(y_sample[0, 0], cmap="jet", vmin=0, vmax=1)
axes[1].set_title("Real Radar")
axes[2].imshow(pred[0, 0], cmap="jet", vmin=0, vmax=1)
axes[2].set_title("Prediction")
plt.tight_layout()
plt.show()
```
